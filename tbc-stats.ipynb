```python
import numpy as np
import seaborn
import pandas
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

palette = seaborn.cubehelix_palette(light=.7, n_colors=6)
seaborn.set_theme(style="darkgrid")
seaborn.set_context("talk", font_scale=1.2)
```

```python
names = ["block_num", "xor_supply", "pswap_supply", "val_supply", "tbc_eth", "tbc_dai", "tbc_val", "tbc_pswap"]
tbcData = np.genfromtxt("tbc-stats.csv", delimiter=',', names=names)
```

```python
blocks = [i[0] for i in tbcData]
xorSupply = [i[1]/1000000000000000000. for i in tbcData]
valSupply = [i[2]/1000000000000000000. for i in tbcData]
pswapSupply = [i[3]/1000000000000000000. for i in tbcData]

tbcEth = [i[4]/1000000000000000000. for i in tbcData]
tbcDai = [i[5]/1000000000000000000. for i in tbcData]
tbcVal = [i[6]/1000000000000000000. for i in tbcData]
tbcPswap = [i[7]/1000000000000000000. for i in tbcData]
```

```python
d = {'xor': xorSupply, "val": valSupply, "pswap": pswapSupply, "tbcEth":tbcEth, "tbcDai":tbcDai, "tbcVal":tbcVal, "tbcPswap":tbcPswap, 'blocks':blocks}
df = pandas.DataFrame(data=d)
```

```python
seaborn.relplot(y="xor", x="blocks", kind="line", data=df, height=9);
plt.savefig('tbc_xorsupply.pdf')
```

```python
seaborn.relplot(y="tbcEth", x="blocks", kind="line", data=df, height=9);
plt.savefig('tbc_eth.pdf')
```

```python
seaborn.relplot(y="tbcDai", x="blocks", kind="line", data=df, height=9);
plt.savefig('tbc_dai.pdf')
```

```python
seaborn.relplot(y="tbcVal", x="blocks", kind="line", data=df, height=9);
plt.savefig('tbc_val.pdf')
```

```python
seaborn.relplot(y="tbcPswap", x="blocks", kind="line", data=df, height=9);
plt.savefig('tbc_pswap.pdf')
```

```python
# network fees
names = ["block_num", "xor_total_fee", "xor_referrer_estimated", "xor_burned_estimated", "xor_dedicated_for_buy_back", "val_burned", "val_reminted_parliament"]
networkFeesDataRaw = np.genfromtxt("network_fees.csv", delimiter=',', names=names, skip_header=1)
```

```python
blocks = [i[0] for i in networkFeesDataRaw]
xor_total_fee = [i[1]/1000000000000000000. for i in networkFeesDataRaw]
xor_referrer_estimated = [i[2]/1000000000000000000. for i in networkFeesDataRaw]
xor_burned_estimated = [i[3]/1000000000000000000. for i in networkFeesDataRaw]

xor_dedicated_for_buy_back = [i[4]/1000000000000000000. for i in networkFeesDataRaw]
val_burned = [i[5]/1000000000000000000. for i in networkFeesDataRaw]
val_reminted_parliament = [i[6]/1000000000000000000. for i in networkFeesDataRaw]
```

```python
networkFeesData = {'blocks': blocks, "xor_total_fee": xor_total_fee, "xor_referrer_estimated": xor_referrer_estimated, "xor_burned_estimated":xor_burned_estimated, "xor_dedicated_for_buy_back":xor_dedicated_for_buy_back, "val_burned":val_burned, "val_reminted_parliament":val_reminted_parliament}
networkFeesDF = pandas.DataFrame(data=networkFeesData)
```

```python
seaborn.relplot(y="val_burned", x="blocks", kind="line", data=networkFeesDF, height=9);
plt.savefig('val_burned.pdf')
```
